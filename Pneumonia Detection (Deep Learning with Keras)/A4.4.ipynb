posterior $\mu$~N(3.437, 0.056)

#### (ii)
n=10, m=3.1, $\bar{x}$=3.5
 
$\tau = \frac{100}{67}$ 
 
$\xi = \frac{1}{10000}$ 
 
$\hat{\mu}=\frac{\xi m+n\tau \bar{x}}{\xi+n\tau} = \frac{\frac{1}{10000}*3.1+10*\frac{100}{67}*3.5}{\frac{1}{10000}+10*\frac{100}{67}} = 3.499997 \approx 3.5$
 
$Var[\mu|x]= \frac{1}{\xi+n\tau} = \frac{1}{\frac{1}{10000}+10*\frac{100}{67}} = 0.067$
 
posterior $\mu$~N(3.5, 0.067)
 
The posterior mean is now $\approx$3.5 which is the same as the sample's mean of 3.5. Compared to the previous 
posterior distribution, the posterior mean has increased and the posterior variance has increase from 0.056 to 0.067. The increases 
in posterior mean and variance are due to the much lower precision in the new prior distribution (compared to the 
previous prior).
